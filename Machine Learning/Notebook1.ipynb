```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('/kaggle/input/quality-prediction-in-a-mining-process/MiningProcess_Flotation_Plant_Database.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.drop(['date'],axis=1,inplace=True)
```

```python
df.columns
```

```python
df.shape
```

```python
for i in list(df.columns):
    df[i]=df[i].str.replace(',','.')
```

```python
for i in list(df.columns):
    df[i]=df[i].astype('float32')
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df= df.drop_duplicates()
df.shape
```

```python
fig=plt.figure(figsize=(18,18))
sns.heatmap(df.corr(),linewidths=0.003,linecolor='white',annot=True)
```

```python
for column in df.columns:
    plt.figure()
    sns.boxplot(y = column, data = df)
    plt.title(f'Boxplot for {column}')
    plt.show()
```

```python
for col in df.columns:
    upper_limit = df[col].quantile(0.95)
    lower_limit = df[col].quantile(0.05)
    df[col] = np.where(df[col] > upper_limit, upper_limit, df[col])
    df[col] = np.where(df[col] < lower_limit, lower_limit, df[col])
```

```python
y=df['% Silica Concentrate']
x=df.drop(['% Silica Concentrate'],axis=1)
```

```python
x.shape , y.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.3)
```

```python
model = LinearRegression()
model.fit(x, y) 

# Calculate the R-squared score on the training data
r_squared = model.score(x, y)
print(f"R-squared score: {r_squared}")
```

```python
poly_feat = PolynomialFeatures(degree =2)
X_train_poly = poly_feat.fit_transform (x_train)
```

```python
poly_model = LinearRegression()
poly_model.fit(X_train_poly, y_train)
```

```python
X_test_poly = poly_feat.transform(x_test)
y_train_poly_predict = poly_model.predict(X_train_poly)
y_test_poly_predict = poly_model.predict(X_test_poly)
```

```python
r2_poly_train = r2_score(y_train,y_train_poly_predict)
r2_poly_test = r2_score(y_test,y_test_poly_predict)
print(r2_poly_train,r2_poly_test)
```

```python
model = DecisionTreeRegressor()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
y_pred_train = model.predict(x_train)
y_pred_test = model.predict(x_test)
```

```python
r2_train = r2_score(y_train, y_pred_train)
r2_test = r2_score(y_test, y_pred_test)
print(f'RÂ² Score for Training Set: {r2_train:.2f}')
print(f'RÂ² Score for Test Set: {r2_test:.2f}')
```
